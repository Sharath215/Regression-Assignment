```python
import pandas as pd 
import xgboost as xg
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset 
```

```python
dataset = pd.get_dummies(dataset, dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_female', 'sex_male','smoker_no', 'smoker_yes']]
independent
```

```python
dependent = dataset[['charges']]
```

```python
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(independent,dependent,test_size = 0.30, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

regressor =GradientBoostingRegressor(random_state=0, n_estimators=100)

regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
r_score
```

```python
import pickle 
filename = "finalized_model_XG_Boost.sav"
pickle.dump(regressor,open(filename,'wb'))
```
